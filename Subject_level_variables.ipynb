```python
import pandas as pd
import numpy as np
from os import path,listdir
```

```python
location = path.join('..','derivatives','main_trial_level.csv')
sheet=pd.read_csv(location) #opening the excel file - trial level data
sheet.head()
```

```python
df = sheet.groupby(["SubjNum","StockValue"]).mean().reset_index()  #groupig things by subjnum and stock value
df.head()
```

we made G10, G2, L10, L2 here:

```python
four_errors = df.pivot(index='SubjNum',columns='StockValue',values='val_estdiff_valid') #pivot it into what we want: each subject has four prob. errors
```

```python
result=sheet.groupby('SubjNum').mean().merge(four_errors, on='SubjNum')
```
