```python
import pandas as pd
import numpy as np
from os import path,listdir
location = path.join('..','derivatives','main_trial_level.csv')
sheet=pd.read_csv(location) #opening the excel file - trial level data
sheet.head()
```

```python
sheet[['stock_chosen','bond_chosen']] = pd.get_dummies(sheet['OptionChosen'])[['stock','bond']]
sheet.head()
```

```python
df = sheet[['SubjNum','TrialNum','stock_chosen','bond_chosen', "Domain"]]
df.head()
```

```python
gain_stock_sum_mean = df[(df['TrialNum'] == 1) & (df['Domain'] == 'GAIN')].groupby('SubjNum').agg({'stock_chosen':['sum','mean']})
loss_stock_sum_mean = df[(df['TrialNum'] == 1) & (df['Domain'] == 'LOSS')].groupby('SubjNum').agg({'stock_chosen':['sum','mean']})
```

```python
gain_stock_frequency = gain_stock_sum_mean[('stock_chosen','mean')]
```

```python
loss_stock_frequency = loss_stock_sum_mean[('stock_chosen','mean')]
```

```python
result=sheet.groupby('SubjNum').mean()
result.head()
```

```python
result['gain_stock_chosen'] = gain_stock_frequency
result['loss_stock_chosen'] = loss_stock_frequency
result.head()
```

```python
result.to_csv("subject_level_risk_taking.csv")
```
