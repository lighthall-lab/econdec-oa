```python
import pandas as pd
import numpy as np
from os import path,listdir
```

```python
location = path.join('..','derivatives','main_trial_level_cleaned.csv')
nback_subject = pd.read_csv(path.join('..','derivatives', 'subjectwise','nback_subject_level.csv'))
sheet=pd.read_csv(location) #opening the excel file - trial level data
sheet.head()
```

```python
df = sheet.groupby(["SubjNum","StockValue"]).mean().reset_index()  #groupig things by subjnum and stock value
df.head()
```

we made G10, G2, L10, L2 here:

```python
four_errors = df.pivot(index='SubjNum',columns='StockValue',values='val_error_3sd_removed') #pivot it into what we want: each subject has four prob. errors
```

```python
result=sheet.groupby('SubjNum', as_index=False).mean().merge(four_errors, on='SubjNum')
```

## Merge main task and nback data

```python
result.rename(columns={'SubjNum': 'sub'}, inplace=True)
main_task_nback = pd.merge(result, nback_subject, on="sub")
```
